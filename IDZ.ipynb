## Обработка изображений. Арифметические операции. Каналы. Вычисление яркости изображения

```python
img = imread('foto.jpg')
```

```python
img.dtype
```

```python
from skimage import img_as_float
```

```python
img_f = img_as_float(img)
```

```python
img.min(), img.max()
```

```python
img_f.min(), img_f.max()
```

```python
imshow(img_f / 1.5)
```

```python
imshow(img_f / 2)
```

```python
imshow(img_f / 4)
```

```python
imshow(img_f / 4 + 0.25)
```

```python
from numpy import clip
```

```python
imshow(clip(img_f * 1.5, 0, 1))
```

```python
imshow(clip(img_f * 2, 0, 1))
```

```python
imshow(clip(img_f * 2 -0.5, 0, 1))
```

```python
img_f = img_as_float(img)
```

```python
r = img_f[:, :, 0]
```

```python
imshow(r)
```

```python
g = img_f[:, :, 1]
```

```python
imshow(g)
```

```python
b = img_f[:, :, 2]
```

```python
imshow(b)
```

```python
from numpy import dstack
```

```python
img_combined = dstack((r, g, b))
```

```python
imshow(img_combined)
```

```python
avg_gray = (r + g + b) / 3
```

```python
imshow(avg_gray)
```
